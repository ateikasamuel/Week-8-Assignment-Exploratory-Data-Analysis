```python
# Import pandas and numpy, DataFrame / Series

import pandas as pd
import numpy as np
import seaborn as sns
from pandas import DataFrame, Series

# Add items for matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# Read the contents of auto-mpg.data into a DataFrame & update the pandas DataFrame columns 
# with the names provided in auto-mpg.names file
mpg = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data',
                  delim_whitespace=True, 
                  names=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name'])

# Display the DataFrame with the updated pandas DataFrame columns
mpg.head()
```

```python
# Replace the ‘?’ in the horsepower column to a pandas-appropriate missing value "NaN"
mpg['horsepower'] = mpg['horsepower'].replace(['?'],'NaN')

# Display the updated DataFrame
mpg.head()
```

```python
# Convert the horsepower column to numeric
mpg['horsepower'] = pd.to_numeric(mpg['horsepower'], errors='coerce').fillna(0).astype('int')

# Display the updated DataFrame
mpg.head()
```

```python
# Convert the column origin to String prior to renaming the values
mpg['origin'] = mpg['origin'].astype('string')

# Convert the values 1,2, & 3 in the origin column to the following values respectively to ‘USA’,‘Asia’, and ‘Europe’
mpg['origin'] = mpg['origin'].replace(to_replace=['1', '2','3'], value=['USA', 'Asia','England'])

# Display the updated DataFrame
mpg
```

```python
# Plot a bar chart that shows the distribution for cylinders using seaborn.

# Loading the dataset
mpg = sns.load_dataset('mpg')

# Crafting and Displaying the bar plot
sns.barplot(x ="cylinders", y ='origin', data = mpg, 
            palette ='plasma')
```

```python
# Plot a scatterplot that shows the relationship between horsepower and weight

# Loading the dataset
mpg = sns.load_dataset('mpg')

# Crafting and Displaying the scatter plot
mpg = sns.regplot(x="horsepower", y="weight", data=mpg, color="g")
```
